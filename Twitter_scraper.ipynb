```python
import pandas as pd
import pickle

import twint
import nest_asyncio
nest_asyncio.apply()

import config
```

```python
with open('data/nyc_candidate_info.csv', 'r') as file:
    candidate_df = pd.read_csv(file)
```

```python
def scrape_user(username):
    c = twint.Config()
    c.Hide_output = True
    c.Username = username
    c.Pandas = True
    c.Limit = 3000
    c.Filter_retweets = False
    
    twint.run.Search(c)
    
    tweets_df = twint.storage.panda.Tweets_df
    
    columns_wanted = ['date', 'tweet']
    tweets_df_filtered = tweets_df[columns_wanted]
    
    with open(f'data/scraped_twitter_profiles/{username}_tweets.pkl', 'wb') as file:
        pickle.dump(tweets_df_filtered, file)
        
    print(f'Done scrapeing {username}!')
    
    return(tweets_df_filtered)
```
