```python
import requests
import dill
from bs4 import BeautifulSoup
import pandas as pd
```

```python
ny_county_politics_url = 'https://www.newyorkcountypolitics.com/2021-election-whos-running-for-what/'

def scrape_page(url):
    _page = requests.get(url)
    _soup = BeautifulSoup(page.text, 'lxml')
    return _soup

soup = scrape_page(ny_county_politics_url)
        
```

```python
def get_candidate_data(html):
    captions = html.select('div.content-inner figure figcaption')
    
    all_candidates = []
    
    for caption in captions:
        parsing = caption.text.split('(')
        if len(parsing) == 2:
            all_candidates.append({'name':parsing[0].strip(), 'party':parsing[1][0]})
        elif len(parsing) == 1:
            all_candidates.append({'name':parsing[0].strip(), 'party':'O'})
    
    return all_candidates

candidates = get_candidate_data(soup)
```

```python
candidate_df = pd.DataFrame(candidates)
candidate_df
```

```python
election_types = []
for count in range(len(candidate_df)):
    if count < 28:
        election_types.append('mayor')
    elif count >= 28 and count < 32:
        election_types.append('public_advocate')
    elif count >= 32 and count < 36:
        election_types.append('comptroller')
    elif count >= 44 and count < 50:
        election_types.append('manhattan_borough_pres')
    else:
        election_types.append('')
        
candidate_df['election_type'] = election_types
candidate_df.head()
```

```python
candidate_df = candidate_df[candidate_df['election_type'] != '']
candidate_df
```
