```python
import pandas as pd
import json
```

```python
#Senate and congress members taken from ProPublica API
#Voting analysis data taken from https://www.govtrack.us/about/analysis#ideology

with open('data/US_congress_data/senate_members.json', 'r') as f:
    senate_json = json.load(f)['results'][0]['members']
senators = pd.DataFrame(senate_json)
    
senate_vote_data = pd.read_csv('data/US_congress_data/senate_analysis.csv')

with open('data/US_congress_data/house_members.json', 'r') as f:
    house_json = json.load(f)['results'][0]['members']
congressmen = pd.DataFrame(house_json)
    
house_vote_data = pd.read_csv('data/US_congress_data/house_analysis.csv')
```

```python
def make_cleaned_df(member_df, analysis_df):
    member_df['full_name'] = member_df['first_name'] + ' ' + member_df['last_name']
    member_df_filtered = member_df[['full_name', 'last_name', 'state', 'party', 'twitter_account']]
    
    analysis_df_filtered = analysis_df[['name', 'ideology', 'leadership', 'description']]
    
    merged_df = member_df_filtered.merge(analysis_df_filtered, left_on='last_name', right_on='name')
    merged_df = merged_df.drop(['last_name', 'name'], axis=1)
    
    return merged_df

senators_df = make_cleaned_df(senators, senate_vote_data)
congressmen_df = make_cleaned_df(congressmen, house_vote_data)
```

```python
senators_df['Chamber'] = 'Senate'
congressmen_df['Chamber'] = 'House'
```

```python
congress_df = pd.concat([senators_df, congressmen_df])
```
