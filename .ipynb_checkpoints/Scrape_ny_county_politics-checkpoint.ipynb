```python
import requests
import dill
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import Request, urlopen
```

```python
url = 'https://www.newyorkcountypolitics.com/2021-election-whos-running-for-what/'

req = Request(url, headers = {'User-Agent': 'Mozilla/5.0'})
webpage = urlopen(req).read()
soup = BeautifulSoup(webpage, 'lxml')
```

```python
def get_candidate_data(candidate):
    
    candidate_dict = {}

    #Getting candidate name and party from figure caption
    cap = candidate.select('figcaption')
    cap_parse = candidate.text.split('(')
    candidate_dict['name'] = cap_parse[0].strip().replace('-', ' ')

    if len(cap_parse) == 2:
        party_letter = cap_parse[1][0]
        if party_letter == 'D':
            candidate_dict['party'] = 'Democrat'
        elif party_letter == 'R':
            candidate_dict['party'] = 'Republican'
    else:
        candidate_dict['party'] = 'Other'

    #Getting candidate website from figure
    candidate_dict['website'] = candidate.find('a')['href']
    
    return candidate_dict

candidates = [get_candidate_data(candidate) for candidate in soup.select('div.content-inner figure')]
```

```python
candidate_df = pd.DataFrame(candidates)
candidate_df.iloc[:40]
```

```python
mayor_df = candidate_df.iloc[:34]
#other_df = candidate_df[candidate_df['election_type'] != 'mayor']
```
