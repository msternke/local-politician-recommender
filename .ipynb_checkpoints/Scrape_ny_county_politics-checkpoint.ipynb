```python
import requests
import dill
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import Request, urlopen
```

```python
url = 'https://www.newyorkcountypolitics.com/2021-election-whos-running-for-what/'

req = Request(url, headers = {'User-Agent': 'Mozilla/5.0'})
webpage = urlopen(req).read()
soup = BeautifulSoup(webpage, 'lxml')
```

```python
def scrape_name_party(cap):
    parsing = caption.text.split('(')
    
    if len(parsing) == 2:
        all_candidates.append({'name':parsing[0].strip(), 'party':parsing[1][0]})
    elif len(parsing) == 1:
        all_candidates.append({'name':parsing[0].strip(), 'party':'O'})
        
    return()

def get_candidate_data(candidate):
    
    candidate_dict = {}

    #Getting candidate name and party from figure caption
    cap = candidate.select('figcaption')
    cap_parse = candidate.text.split('(')
    candidate_dict['name'] = cap_parse[0].strip()

    if len(cap_parse) == 2:
        candidate_dict['party'] = cap_parse[1][0]
    else:
        candidate_dict['party'] = 'O'

    #Getting candidate website from figure
    candidate_dict['website'] = candidate.find('a')['href']
    
    return candidate_dict

candidates = [get_candidate_data(candidate) for candidate in soup.select('div.content-inner figure')]
```

```python
candidate_df = pd.DataFrame(candidates)
candidate_df.iloc[50:80]
```

```python
elections = ['mayor']*32 + ['public_advocate']*4 + ['comptroller']*6 + ['manhattan_DA']*8 + \
['manhattan_borough_pres']*6 + ['district_1']*7 + ['district_2']*1 + ['district_3']*6 + \
['district_4']*1 + ['district_5']*7 + ['district_6']*6 + ['district_7']*11 + ['district_8']*2 + \
['district_9']*11 + ['district_10']*5

candidate_df['election_type'] = elections
candidate_df
```

```python
mayor_df = candidate_df[candidate_df['election_type'] ==  'mayor']
other_df = candidate_df[candidate_df['election_type'] != 'mayor']
```
